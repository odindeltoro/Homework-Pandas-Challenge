```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
totalFemale =dfPurchaseData["Gender"] == "Female"
totalMale =dfPurchaseData["Gender"] == "Male"
totalOther =dfPurchaseData["Gender"] == "Other / Non-Disclosed"
femaleCount =dfPurchaseData.loc[totalFemale,:]["SN"].nunique()
maleCount =dfPurchaseData.loc[totalMale,:]["SN"].nunique()
otherCount =dfPurchaseData.loc[totalOther,:]["SN"].nunique()
femalePurchaseDatadf = dfPurchaseData.loc[totalFemale, :]
malePurchaseDatadf = dfPurchaseData.loc[totalMale, :]
otherPurchaseDatadf = dfPurchaseData.loc[totalOther, :]

# Determine Purchase Count
femalePurchaseCount = femalePurchaseDatadf["Purchase ID"].count()
malePurchaseCount = malePurchaseDatadf["Purchase ID"].count()
otherPurchaseCount = otherPurchaseDatadf["Purchase ID"].count()

# Total Purchase Value
femaleTotalPurchaseValue = femalePurchaseDatadf["Price"].sum()
maleTotalPurchaseValue = malePurchaseDatadf["Price"].sum() 
otherTotalPurchaseValue = otherPurchaseDatadf["Price"].sum()

# Calculate Average Purchase Price
femalePurchaseAveragePrice = round(femaleTotalPurchaseValue / femalePurchaseCount, 2)
malePurchaseAveragePrice = round(maleTotalPurchaseValue / malePurchaseCount, 2)
otherPurchaseAveragePrice = round(otherTotalPurchaseValue / otherPurchaseCount, 2)

# Calculate Average Purchase Total per Person by Gender
femalePurchaseTotal = round(femaleTotalPurchaseValue / femaleCount, 2)
malePurchaseTotal = round(maleTotalPurchaseValue / maleCount, 2)
otherPurchaseTotal = round(otherTotalPurchaseValue / otherCount, 2)
```

```python
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame
femalePurchaseAveragePrice = "${:,.2f}".format(femalePurchaseAveragePrice)
malePurchaseAveragePrice = "${:,.2f}".format(malePurchaseAveragePrice)
otherPurchaseAveragePrice = "${:,.2f}".format(otherPurchaseAveragePrice)
femaleTotalPurchaseValue = "${:,.2f}".format(femaleTotalPurchaseValue)
maleTotalPurchaseValue = "${:,.2f}".format(maleTotalPurchaseValue)
otherTotalPurchaseValue = "${:,.2f}".format(otherTotalPurchaseValue)
femalePurchaseTotal = "${:,.2f}".format(femalePurchaseTotal)
malePurchaseTotal = "${:,.2f}".format(malePurchaseTotal)
otherPurchaseTotal = "${:,.2f}".format(otherPurchaseTotal)

genderPurchasingAnalysis = pd.DataFrame({"Purchase Count": [femalePurchaseCount, malePurchaseCount, otherPurchaseCount],"Average Purchase Price": [femalePurchaseAveragePrice, malePurchaseAveragePrice, otherPurchaseAveragePrice],"Total Purchase Value": [femaleTotalPurchaseValue, maleTotalPurchaseValue, otherTotalPurchaseValue],"Average Purchase Total per Person": [femalePurchaseTotal, malePurchaseTotal, otherPurchaseTotal]},index=["Female", "Male", "Other / Non-Disclosed"])
genderPurchasingAnalysis
```

```python
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
# Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
spenderPurchaseTabledf["Purchase Count"] = spenderPurchaseTabledf["Purchase Count"]
spenderPurchaseTabledf["Average Purchase Price"] = round(spenderPurchaseTabledf["Average Purchase Price"],2)
spenderPurchaseTabledf["Total Purchase Value"] = round(spenderPurchaseTabledf["Total Purchase Value"],2)
topSpenderPurchasedf = spenderPurchaseTabledf.nlargest(5, ["Total Purchase Value"])
topSpenderPurchasedf
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Create a summary data frame to hold the results


* Sort the purchase count column in descending order

```python
# Results of Popular Items
```

```python
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value
popularItemsTabledf = dfPurchaseData[["Item ID","Item Name", "Price"]]
popularItemsdf = popularItemsTabledf.groupby(["Item ID","Item Name"]).agg({"Price":["count", "mean", "sum"]})
popularItemsdf.columns = ["Purchase Count","Item Price" , "Total Purchase Value"]
popularItemsdf
```

```python
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
# Identify the 5 most popular items by purchase count, then list (in a table):
popularItemsdf["Purchase Count"] = popularItemsdf["Purchase Count"]
popularItemsdf["Item Price"] = round(popularItemsdf["Item Price"],2)
popularItemsdf["Total Purchase Value"] = round(popularItemsdf["Total Purchase Value"],2)
topPopularItemsdf = popularItemsdf.nlargest(10, ["Purchase Count"])
topPopularItemsdf
```
